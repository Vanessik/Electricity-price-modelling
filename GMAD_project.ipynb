## Locally linear projection

Locally linear projection (LLP) is recommended to smooth the manifold and reduce the noise.

```
from sklearn.neighbors import NearestNeighbors

def LLP(Y, k_neighb=23, k0=4):
  smooth_prices = []
  Y_shape = Y.shape[0]
  neigh = NearestNeighbors(n_neighbors=k_neighb)
  neigh.fit(traj_price)
  #collect neighbors
  neighb = neigh.kneighbors(traj_price, n_neighbors=k_neighb, return_distance=False)
  for i in range(Y_shape):
    #for each sample yi
    #compute SVD and devine subspace
    U, _, _ = np.linalg.svd(traj_price[neighb[i]].T, full_matrices = True, compute_uv = True)
    A = U[:, :k0].copy()
    smooth_prices.append(projection_vec_A(Y[i], A))
  return np.array(smooth_prices)


def projection_vec_A(vector, A):
  '''Formula for projection taken from here 
  https://en.wikibooks.org/wiki/Linear_Algebra/Projection_Onto_a_Subspace 
  '''
  return A.dot(np.linalg.inv(A.T.dot(A))).dot(A.T).dot(vector)
```

## LLE Reconstruction

 - Implement LLE reconstruction algorithm from the article to reconstruct high dimension data from low dimension.

```
from sklearn.manifold.locally_linear import barycenter_kneighbors_graph

def LLE_reconstruction(Z, Y, X, idx_=-1, k=23, new=False):
  '''
  --------
  Input:

  Z: forecast array;
  Y: low dim array after LLE;
  X: initial high dim array;
  idx_: point of reconstruction;
  k: num of neighbors;
  new: points either forecasted or in Y;
  --------
  Returns:
  Reconstracted high dimension point
  '''
  if new:
    concatenated = np.concatenate((Y, Z), axis=0)
  else:
    concatenated = Y.copy()
  _, idx = barycenter_kneighbors_graph(concatenated , 23)[idx_].nonzero()
  weights = barycenter_kneighbors_graph(concatenated , 23)[idx_].toarray()[0][idx]
  reconstructed = np.array([X[idx[i]]*weights[i] for i in range(idx.shape[0])]).sum(axis=0)
  return reconstructed
```

Implement weekly average prediction error of one-week ahead prediction and daily average prediction error of one-day ahead prediction by formulas from article:

```
def WPE_d(predicted, actual):
  x_ = np.abs(actual).sum()/7/24
  return np.abs(actual-predicted).sum()/x_/7/24

def WPE_w(predicted, actual):
  x_ = np.abs(actual).sum(axis=0)/7/24
  return 1/7/7/24*(np.abs(actual-predicted).sum(axis=0)/x_).sum()
```

```
WPEd_holt = WPE_d(result['LBMP ($/MWHr)'].values.reshape(731, 24).copy()[-1], reconstr_holt_day)
WPEd_holt_log = WPE_d(traj_price_log[-1], reconstr_holt_day_log)
WPEw_holt = WPE_w(result['LBMP ($/MWHr)'].values.reshape(731, 24).copy()[-7:], reconstr_holt_week)
WPEw_holt_log = WPE_w(traj_price_log[-7:], reconstr_holt_week_log)

print('Holt Winter WPEd loss:', WPEd_holt)
print('Holt Winter WPEd loss for log data:', WPEd_holt_log)
print('Holt Winter WPEw loss:', WPEw_holt)
print('Holt Winter WPEw loss for log data:', WPEw_holt_log)
```

## 2) Vector AR model

```
from statsmodels.tsa.api import VAR
```

```
def VAR_model(data, points=1, plot=False):
  ''' 
  -------
  Inputs:
  points: number to forecast;
  data: obtained after LLE;
  -------
  Returns:
  forecast defined number of points.
  '''
  model = VAR(data)
  results = model.fit(maxlags=20, ic='aic')
  lag_order = results.k_ar
  if plot:
    results.plot_forecast(points);
  return results.forecast(data[-lag_order:].values, points)
```

Define d, p and q parameters for ARIMA.

The value of d  is the minimum number of differencing needed to make the series stationary. And if the time series is already stationary, then d = 0.

p is the order of the ‘Auto Regressive’ (AR) term. It refers to the number of lags of Y to be used as predictors. And q is the order of the ‘Moving Average’ (MA) term. It refers to the number of lagged forecast errors that should go into the ARIMA Model.

```
def ARIMA_model(train, test, plot=False): 
  model = ARIMA(train, order=(p, d, q))  
  fitted = model.fit(disp=-1)  
  # Forecast
  fc, se, conf = fitted.forecast(test.shape[0], alpha=0.05)  # 95% conf
  # Make as pandas series
  fc_series = pd.Series(fc, index=test.index)
  lower_series = pd.Series(conf[:, 0], index=test.index)
  upper_series = pd.Series(conf[:, 1], index=test.index)
  # Plot
  if plot:
    plt.figure(figsize=(12,5), dpi=100)
    #plt.plot(train[500-test.shape[0]:], label='training')
    plt.plot(test, label='actual')
    plt.plot(fc_series, label='forecast')
    plt.fill_between(lower_series.index, lower_series, upper_series, 
                    color='k', alpha=.15)
    plt.title('Forecast vs Actuals')
    plt.legend(loc='upper left', fontsize=8)
    plt.show()
  return fc_series.values
```

```
forecast_ARIMA_day = ARIMA_model(train_day, test_day, plot=True)
```
